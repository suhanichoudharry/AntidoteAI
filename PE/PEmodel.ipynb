```python
import os
import pandas
import numpy
import pickle
import pefile
import sklearn.ensemble as ek
from sklearn.feature_selection import SelectFromModel
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import svm
import sklearn.metrics as metrics
```

```python
# important feature selection

extratrees = ek.ExtraTreesClassifier().fit(X,y)             # Classifier with data
selector = SelectFromModel(extratrees,prefit=True)          # It will select important features
X_new = selector.transform(X)                               # selected features from X
n_features = X_new.shape[1]                                 # count of X_new
```

```python
n_features
```

```python
# print important features

features = []
index = numpy.argsort(extratrees.feature_importances_)[::-1][:n_features]    # features index

print(index)
# index array contains the indices of most important features in descending order of importance
```

```python
# print selected features with importance


for f in range(n_features):
    print("%d. feature %s (%f)" % (f + 1, dataset.columns[2+index[f]], extratrees.feature_importances_[index[f]])) # +2 cuz we removed md5,name
    features.append(dataset.columns[2+f]) 
```

```python
# splitting the data (70:30)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.30, stratify = y)
```
